```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Create sample e-commerce data
np.random.seed(42)  # For reproducible results

# Generate 1000 customer records
data = {
    'customer_id': range(1000),
    'purchase_amount': np.random.normal(100, 30, 1000),  # Normal distribution around $100
    'age': np.random.randint(18, 70, 1000),
    'items_purchased': np.random.randint(1, 10, 1000),
    'time_spent_mins': np.random.normal(30, 10, 1000),
    'is_returning': np.random.choice([0, 1], 1000, p=[0.7, 0.3]),  # 30% are returning customers
    'satisfaction_score': np.random.randint(1, 6, 1000)  # 1-5 rating
}

# Create DataFrame
df = pd.DataFrame(data)

# 1. Basic Data Exploration
print("\n--- Basic Data Information ---")
print(df.info())  # Shows data types and null values
print("\n--- Summary Statistics ---")
print(df.describe())  # Shows statistical summary

# 2. Data Cleaning
# Remove outliers from purchase_amount (those beyond 3 standard deviations)
purchase_mean = df['purchase_amount'].mean()
purchase_std = df['purchase_amount'].std()
df_cleaned = df[
    (df['purchase_amount'] >= purchase_mean - 3*purchase_std) &
    (df['purchase_amount'] <= purchase_mean + 3*purchase_std)
]

# 3. Customer Segmentation Analysis
# Calculate average metrics per customer
customer_segments = df_cleaned.groupby('is_returning').agg({
    'purchase_amount': 'mean',
    'items_purchased': 'mean',
    'satisfaction_score': 'mean'
}).round(2)

print("\n--- Customer Segments Analysis ---")
print(customer_segments)

# 4. Visualization
# Create a figure with multiple subplots
plt.figure(figsize=(15, 10))

# Plot 1: Purchase Amount Distribution
plt.subplot(2, 2, 1)
sns.histplot(data=df_cleaned, x='purchase_amount', bins=30)
plt.title('Distribution of Purchase Amounts')
plt.xlabel('Purchase Amount ($)')
plt.ylabel('Count')

# Plot 2: Age vs Purchase Amount
plt.subplot(2, 2, 2)
sns.scatterplot(data=df_cleaned, x='age', y='purchase_amount')
plt.title('Age vs Purchase Amount')
plt.xlabel('Age')
plt.ylabel('Purchase Amount ($)')

# Plot 3: Satisfaction Score Distribution
plt.subplot(2, 2, 3)
sns.countplot(data=df_cleaned, x='satisfaction_score')
plt.title('Distribution of Satisfaction Scores')
plt.xlabel('Satisfaction Score')
plt.ylabel('Count')

# Plot 4: Average Purchase Amount by Satisfaction Score
avg_purchase_by_satisfaction = df_cleaned.groupby('satisfaction_score')['purchase_amount'].mean()
plt.subplot(2, 2, 4)
avg_purchase_by_satisfaction.plot(kind='bar')
plt.title('Average Purchase Amount by Satisfaction Score')
plt.xlabel('Satisfaction Score')
plt.ylabel('Average Purchase Amount ($)')

plt.tight_layout()
plt.show()

# 5. Advanced Analysis: Correlation Study
correlation_matrix = df_cleaned[['purchase_amount', 'age', 'items_purchased', 
                               'time_spent_mins', 'satisfaction_score']].corr()

print("\n--- Correlation Analysis ---")
print(correlation_matrix.round(2))

# 6. Customer Insights
# Calculate key metrics
avg_purchase = df_cleaned['purchase_amount'].mean()
most_common_age = df_cleaned['age'].mode()[0]
satisfaction_rate = (df_cleaned['satisfaction_score'] >= 4).mean() * 100

print("\n--- Key Business Insights ---")
print(f"Average purchase amount: ${avg_purchase:.2f}")
print(f"Most common customer age: {most_common_age}")
print(f"Percentage of satisfied customers (score ≥ 4): {satisfaction_rate:.1f}%")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Create sample dataset
np.random.seed(42)
n_samples = 1000

# Create a DataFrame with multiple variables
df = pd.DataFrame({
    'age': np.random.normal(30, 10, n_samples),
    'income': np.random.normal(60000, 15000, n_samples),
    'education_years': np.random.normal(16, 3, n_samples),
    'job_satisfaction': np.random.randint(1, 6, n_samples),
    'department': np.random.choice(['Sales', 'Engineering', 'Marketing', 'HR'], n_samples),
    'experience': np.random.normal(8, 4, n_samples)
})

# Set the style for all plots
sns.set_style("whitegrid")

# 1. Distribution Plot (combines histogram and KDE)
plt.figure(figsize=(12, 8))
plt.subplot(2, 2, 1)
sns.histplot(data=df, x='age', kde=True)
plt.title('Distribution Plot (Histogram with KDE)')

# 2. Box Plot with categorical variable
plt.subplot(2, 2, 2)
sns.boxplot(data=df, x='department', y='income')
plt.title('Box Plot by Department')
plt.xticks(rotation=45)

# 3. Violin Plot
plt.subplot(2, 2, 3)
sns.violinplot(data=df, x='department', y='job_satisfaction')
plt.title('Violin Plot of Job Satisfaction')
plt.xticks(rotation=45)

# 4. Scatter Plot with regression line
plt.subplot(2, 2, 4)
sns.regplot(data=df, x='experience', y='income')
plt.title('Scatter Plot with Regression Line')

plt.tight_layout()
plt.show()

# 5. Heatmap of correlation matrix
plt.figure(figsize=(10, 8))
correlation_matrix = df.select_dtypes(include=[np.number]).corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', center=0)
plt.title('Correlation Heatmap')
plt.show()

# 6. Pair Plot (shows relationships between multiple variables)
sns.pairplot(df.select_dtypes(include=[np.number]))
plt.show()

# 7. Count Plot (for categorical data)
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='department')
plt.title('Count Plot of Departments')
plt.show()

# 8. Joint Plot (combines scatter plot and distributions)
sns.jointplot(data=df, x='age', y='income', kind='hex')
plt.show()
```
